```python











































































































































































































































































































































































































# outside of the class
class std:
    college='JNM college'
# before modification
print(std.college)
# after modification
std.college='SBM college'
print(std.college)
```

```python
# inside constructor
class std:
    clg='JNM college'
    def __init__(self):
        std.clg='SBM clg'
print(std.clg)
# if u exe without object creation then it will give you the old output
# because without obj the constructor will not execute
s1=std()
print(std.clg)
```

```python
# inside instance method
class std:
    clg='JNM college'
    def create(self):
        std.clg='SBM college'
print(std.clg)

s1=std()
s1.create()

print(std.clg)
```

```python
# inside class method
# you can also use cls variable
class std:
    clg='JNM clg'
    @classmethod
    def cm(cls):
        std.clg="alpha college"
print(std.clg)
# s1=std()
std.cm()
print(std.clg)
```

```python
# inside static method 
class std:
    clg='JNM clg'
    @staticmethod
    def sm():
        std.clg='alpha colllllege'
print(std.clg)
std.sm()
print(std.clg)
```

```python
# by using self / referance variable
class std:
    clg ='NIT clg'
    def __init__(self) :
        self.clg='MM college' #you can't modify a static variable by using self becoz it will create an instance variable in the place of modification
print(std.clg)
s1=std()
print(std.clg)
# print(std.__dict__)
print(s1.__dict__)

```

```python
# using referance  variable
class std:
    clg ='NIT clg'
s1=std()
s1.clg='bkw college' #again it will also create an instance variable 
print(std.clg)
print(s1.clg)
```
